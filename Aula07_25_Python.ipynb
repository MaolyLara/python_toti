● Caso o aluno tenha que fazer o trabalho final, a média final é calculada da seguinte forma: a média aritmética entre a nota do trabalho final e a média m anterior.

* Se mf for maior ou igual a 5.0 o aluno está aprovado, se for menor o
aluno está reprovado.

● Escreva um programa em python que calcule a nota final desse aluno.

```python
def calcula_media(valor1, valor2):
  return (valor1 + valor2)/2

t1 = float(input("Entre com a nota do trabalho 1 "))
t2 = float(input("Entre com a nota do trabalho 2 "))

m = calcula_media(t1, t2) # m é a média dos trabalhos

if m >= 7.0: # aluno aprovado
  print("Aluno aprovado! Com média: ", m)
elif m >= 3.0 and m < 7.0: # aluno em situação de trabalho final
  tf = float(input("Aluno em situação de Trabalho Final. Entre a nota do TF "))
  mf = calcula_media(m, tf)
  if mf >= 5.0:
    print("Aluno aprovado com a média ", mf)
  else: #if ANINHADO
    print("Aluno reprovado com a média ", mf)
else: # aluno reprovado direto
  print("Aluno reprovado direto com média ", m)
```

**EXERCÍCIO - WHILE**

● Escrever um programa que leia quantos ‘r’ ou ‘R’ tenha na frase “o rato roeu a roupa do Rei de Roma”


```python
def leia(frase, contador):
  contador = 0 
  # contador de R ou r, comença no 0 pq não tem nenhum no inicio 
  i = 0 
  # i para fazer controle do index
  while i < len(frase):
    if frase[i] == 'r' or frase[i] == 'R':
      contador = contador + 1
    i = i + 1
  print(f'Tem {contador} r ou R')
```

```python
contador = 0
frase= input('Ingrese a frase: ')
leer = leia(frase, contador)
```

**Listas nos parâmetros de Função**

---

**Formas de passagem de Parâmetro**

1.   Por valor: o valor da variável na chamada é copiado para a variável da função, assim as alterações **não são** refletidas na variável original

2.   Por referência: como se a própria variável fosse chamada, as alterações **refletem** na variável original

No Python, em geral, a passagem é feita por valor. Entretnato, quandoe estamos falando de listas a passagem é por referência

```python
# Por valor

def fazConta(a, b):
  a = 100
  b = a + 1
  return a + b

a = 1
b = 2
print(a)
print(b)
print(fazConta(a, b))
print(a)
print(b)
```

```python
# Por referência

def fazConta(lista):
  lista[0] = 3
  return lista[0] + lista[1]

lista = [1, 2]
print(lista)
print(fazConta(lista))
print(lista)
```

Solução

Se não quero perder a lista original, passo uma cópia para a função

```python
def fazConta(lista):
  lista[0] = 3
  return lista[0] + lista[1]

lista = [1, 2]
print(lista)
print(fazConta(lista.copy())) # ou faço lista2 = lista.copy() e passo lista2
print(lista)
```
