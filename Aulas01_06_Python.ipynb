 **Index**

 ● Posso querer pegar um elemento da string

  *  Um caractere

    ■ Uma letra, por exemplo

**● `string[index]`**

● Index é um número

  * Posição na string

  image.png

````python
# primeira posiçao

elemento = 'posição do elemento em uma string'
primer_elemento = elemento [0]

print(primer_elemento)
````

````python
#terceira posição

elemento = 'posição do elemento em uma string'
tercer_elemento = elemento [3]

print(tercer_elemento)
````

````python
#terceira posição com conhecimento do cumprimento da string

elemento = 'posição do elemento em uma string'
ultimo_elemento = elemento [32]

print(ultimo_elemento)
````

````python
#terceira posição sem conhecimento do cumprimento da string - usando o tamnho da string

elemento = 'posição do elemento em uma string'
index_ultimo_elemento = len(elemento) - 1

ultimo = elemento[index_ultimo_elemento]

print(ultimo)
````

**Index Negativo**



````python
# Index negativo

elemento = 'posição do elemento em uma string'
ultimo_elemento = elemento [-1]

print(ultimo_elemento)
````

**SLICING**

● Posso querer pegar substrings

  * Palavras dentro de frases

  * Letras dentro de palavras...

● **`string[início:fim]`**

● A substring pega do elemento início e vai até o elemento

image.png

````python
#Quebrando em substrings

elemento = 'posição do elemento em uma string'
sub_elemento = elemento [0:2]

print(sub_elemento)
````

````python
#Quebrando em substrings

elemento = 'posição do elemento em uma string'
sub_elemento = elemento [0:3]

print(sub_elemento)
````

````python
'''#Pegando a palavra elemento, considerando que el cumprimento da string é de 33 elementos, 
de esquerda a dereita a posição de elemento é 11 inicio até o 18, para crumpria a regla do slicing até menos um elemento no caso a expreção seria [11:19]'''

elemento = 'posição do elemento em uma string'

cumprimento= len(elemento)
print(cumprimento)

sub_elemento = elemento [11:19]

print(sub_elemento)
````

**ACHAR ELEMENTOS**

● Posso querer achar caracteres em uma string

  * find

● `string.find(o que quero achar)`

● O resultado dessa operação find é a posição do elemento
  * O index

````python
#Como achar um elemento, "Só vai informar a posição do primer elemento"

frase = 'Maoly Lara Serrano'

pos = frase.find("a")

print(pos)
````

````python
# Si no encontrar o elemento retorna -1

frase = 'Maoly Lara Serrano'

pos = frase.find("Y") #Python diferencia maiculas e minisculas

print(pos)
````
